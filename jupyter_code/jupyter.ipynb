```python
import cv2
import numpy as np
from ultralytics import YOLO

# ── 1. Load YOLO11 model ──────────────────────────────────────────────────────
model = YOLO("yolo11n.pt")  # or yolo11s.pt / yolo11m.pt for better accuracy

# ── 2. Load image ─────────────────────────────────────────────────────────────
image_path = "Segi Fresh fish 1.jpg"
img = cv2.imread(image_path)
img_display = img.copy()

# ── 3. Run YOLO11 inference ───────────────────────────────────────────────────
results = model(img, conf=0.4, classes=[0])  # class 0 = person only

# ── 4. Green color range in HSV (tuned for DARK green like Segi uniform) ──────
LOWER_GREEN = np.array([40,  40,  20])   # H, S, V - lower bound
UPPER_GREEN = np.array([85, 255, 110])   # H, S, V - upper bound
GREEN_THRESHOLD = 0.08                   # min 8% green pixels in torso region

# ── 5. Process each detected person ──────────────────────────────────────────
def is_wearing_green(person_crop):
    """Check if cropped person region contains significant green clothing."""
    if person_crop.size == 0:
        return False, 0.0

    # Focus on torso (top 25%–65% of bounding box = shirt area)
    h, w = person_crop.shape[:2]
    torso = person_crop[int(h * 0.25): int(h * 0.65), :]

    hsv = cv2.cvtColor(torso, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, LOWER_GREEN, UPPER_GREEN)

    green_ratio = np.sum(mask > 0) / mask.size
    return green_ratio >= GREEN_THRESHOLD, round(green_ratio * 100, 2)


detected_count = 0

for result in results:
    boxes = result.boxes
    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = float(box.conf[0])

        # Crop the detected person
        person_crop = img[y1:y2, x1:x2]
        has_green, green_pct = is_wearing_green(person_crop)

        if has_green:
            detected_count += 1

            # ── Draw bounding box (green) ──────────────────────────────────
            cv2.rectangle(img_display, (x1, y1), (x2, y2), (0, 200, 0), 3)

            # ── Label ─────────────────────────────────────────────────────
            label = f"Green Shirt | {green_pct}% | conf:{conf:.2f}"
            label_size, _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.55, 2)
            cv2.rectangle(img_display,
                          (x1, y1 - 25), (x1 + label_size[0] + 8, y1),
                          (0, 200, 0), -1)
            cv2.putText(img_display, label,
                        (x1 + 4, y1 - 7),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.55,
                        (255, 255, 255), 2)

        else:
            # Draw regular person box (grey, no label)
            cv2.rectangle(img_display, (x1, y1), (x2, y2), (180, 180, 180), 1)

# ── 6. Summary overlay ────────────────────────────────────────────────────────
summary = f"Green Shirt Staff Detected: {detected_count}"
cv2.rectangle(img_display, (10, 10), (420, 45), (0, 0, 0), -1)
cv2.putText(img_display, summary, (15, 35),
            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 100), 2)

# ── 7. Save & display result ──────────────────────────────────────────────────
output_path = "output_green_shirt_detected.jpg"
cv2.imwrite(output_path, img_display)
print(f"[✓] Saved: {output_path}")
print(f"[✓] Green shirt persons found: {detected_count}")

cv2.imshow("YOLO11 - Green Shirt Detection", img_display)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2 

import numpy as np 

from ultralytics import YOLO 

 

model = YOLO('yolo11n.pt')  

image_path = 'Segi Fresh testing.jpg' 

img = cv2.imread(image_path) 

 

# 1. Convert the entire image to HSV for color checking 

hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) 

 

# 2. Define the "Green Shirt" HSV range 

# These ranges can be tuned; 35-85 covers most greens 

lower_green = np.array([35, 40, 40])  

upper_green = np.array([85, 255, 255]) 

 

results = model.predict(source=image_path, conf=0.25) 

 

for result in results: 

    for box in result.boxes: 

        if int(box.cls[0]) == 0:  # Only check "Person" 

            x1, y1, x2, y2 = map(int, box.xyxy[0]) 

             

            # 3. Create a mask for just this person's area 

            person_roi_hsv = hsv_img[y1:y2, x1:x2] 

            mask = cv2.inRange(person_roi_hsv, lower_green, upper_green) 

             

            # 4. Calculate how much green is in the box 

            green_pixel_count = np.sum(mask > 0) 

             

            # Only circle if they have a significant amount of green (e.g., > 500 pixels) 

            if green_pixel_count > 500: 

                center_x, center_y = (x1 + x2) // 2, (y1 + y2) // 2 

                radius = (y2 - y1) // 3 

                cv2.circle(img, (center_x, center_y), radius, (0, 255, 0), 3) 

                cv2.putText(img, "Target Identified", (x1, y1-10),  

                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2) 

 

cv2.imwrite('hsv_green_detected.jpg', img) 

cv2.imshow('Green Shirt Filter', img) 

cv2.waitKey(0) 

cv2.destroyAllWindows() 
```
